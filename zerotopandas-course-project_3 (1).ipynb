# Most IPL Matches played in a Venue

```python
new_match_raw_df.groupby(['venue','id']).count().droplevel(level=1).index.value_counts()
```

#  IPL Matches Played by Each Team

```python
plt.figure(figsize=(10,8))
data = new_ball_df['bowling_team'].value_counts().sort_values(ascending=False)
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('Matches Played')
plt.ylabel('Team')
plt.show()
```

Above graph show that most ipl played by mumbai indian and followed by royal challenger bangalouru ,kolkata night riders so on and least match played by kochi tuskers kerla 
which prsent time this not a team  it removed from upcoming ipl season

# Most Run Scored by IPL Teams

```python
new_ball_df.groupby(['batting_team'])['total_runs'].sum().sort_values(ascending=False)
```

To calculate the most run scored by a team across all seasons we have grouped by Team and have summed up the run scored by them. And finally, sort them in descending order.

# Top 5 Players with the Highest Number of Man of the Match Awards

```python
temp_data=new_match_raw_df['player_of_match'].value_counts().head()
print(temp_data)
sns.barplot(x=temp_data.index,y=temp_data.values,data=new_match_raw_df)
plt.title("Top 5 MoM")
plt.xticks(rotation=90)
plt.xlabel("Match Count")
plt.ylabel("Player")
plt.show()
```

# Most Boundary (4s) hit by a Batsman

```python
plt.figure(figsize=(10,8))
data = new_ball_df[new_ball_df['batsman_runs'] == 4]['batsman'].value_counts()[:10]
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('Fours')
plt.ylabel('Batsman')
plt.show()
```

we oserve that The Indian Gabbar, Shikhar Dhawan is at the top of the list with more than 600 boundaries followed by Virat Kohli and David warner.

# The Bowler with Team-wise Performance

```python
mask=new_ball_df['bowler']=='PP Chawla'
new_ball_df[mask].groupby('batting_team')['total_runs'].agg('sum').plot(kind='bar');
```

We are taking the example of PP Chawla. This bowler has given the highest number of runs in the IPL history till 2020. We summed up the total runs given by PP Chawla to the opponent team.

It’s clear that if you have PP Chawla in your team then don’t let him play against MI, CSK, RCB, RR, and DC.

# Over-wise Batting Performance of Each Team in the IPL (2008–2020)

```python
data=new_ball_df[mask]
data=data[['batting_team','over','batsman_runs']]
x=data.pivot_table(values='batsman_runs', index='batting_team', columns='over', aggfunc='count')
sns.heatmap(x, cmap='summer');
```

if you are playing against MI or CSK, then you have to play with your best bowling attack line-up from the first over. MI’s batsmen are silent in the second and third over, after that they go on rampage mode against their opponent. The same goes for CSK and RCB too.

## Asking and Answering Questions

Here are some interesting question arise with the help of visulisation of grap interesting fact about batsman as well as bowler in the whole season ipl 2008-2020



#### Q1: Which bowler deleveries in most dot ball

```python
plt.figure(figsize=(10,8))
data = new_ball_df[new_ball_df['total_runs'] == 0].groupby('bowler').count()['id'].sort_values(ascending=False)[:10]
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('Dot Balls')
plt.ylabel('bowler')
plt.show()
```

The Indian bowler Harbhajan Singh has bowled the most number of Dot balls followed by R. Ashwin and Bhuvneshwar Kumar

#### Q2: which team give most extra run in whole season?

```python
plt.figure(figsize=(10,8))
data =new_ball_df.groupby(['batting_team'])['extra_runs'].agg('sum').sort_values(ascending=False)
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('Runs')
plt.ylabel('Teams')
plt.show()
```

Mumbai Indians have given the most number of extras (byes, no balls, wides) followed by Kolkata Knight Riders and Kings XI Punjab.

#### Q3: find the top 10 bowler which give highest number of  run?

```python
new_ball_df.groupby('bowler')['total_runs'].agg('sum').reset_index().sort_values('total_runs', ascending=False).head(10)
```

here we see that pp chawla give most run in ipl history followed by harbhajan sing so both player from chennai super king team so in some match both payeer had not played in aginst some team because he give most run in each ball

#### Q4: find the player name which have most ipl century ?

```python
runs = new_ball_df.groupby(['batsman','id'])['batsman_runs'].sum()
runs[runs >= 100].droplevel(level=1).groupby('batsman').count().sort_values(ascending=False)[:10]
```

The Universe Boss Chris Gayle is at the top of the list in scoring the most number of centuries in IPL history. He has hit six tons and has scored 4804 runs in IPL.

His former teammate Virat Kohli has scored five hundred’s and he is at the second spot in the list followed by Watson, AB de Villiers, Brendon McCullum, and David Warner
